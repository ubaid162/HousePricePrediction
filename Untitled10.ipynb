### Removing unnecessary columns

```python
data.drop(['balcony','society','availability'],axis = 1,inplace = True)
data
```

### Removing Duplicates

```python
data.duplicated().sum()
```

```python
data.drop_duplicates()
```

### Handling missing values

```python
data.isna().sum()
```

```python
data['bath'] = data['bath'].fillna(method = 'ffill')
```

```python
data['location'].fillna(data['location'].mode()[0],inplace =True)
```

```python
data['size'].fillna(data['size'].mode()[0],inplace = True)
```

```python
data['bhk'] = data['size'].str.extract('(\d+)').astype('int')
```

```python
data
```

```python
data.drop('size',axis =1,inplace = True)
```

```python
data.info()
```

```python
def convert_to_float(val):
    if isinstance(val,str) and '-' in val:
        start,end = val.split('-')
        return (float(start) +float(end))/2
    try:
        return float(val)
    except:
        return None

data['total_sqft'] = data['total_sqft'].apply(convert_to_float)
```

```python
data.dropna(inplace = True)
```

```python
#Addding Extra column
data['price_per_sqft'] = data['price']*100000/data['total_sqft']
```

### Feature Encoding

```python
# Removing less no of location to reduce noice
loc_count = data['location'].value_counts()
loc_less_than_10 = loc_count[loc_count <= 10].index
data['location'] = data['location'].apply(lambda x : 'other' if x in loc_less_than_10 else x)
```

```python
# Rempve where sqrt of bhk too small unlealistic
data = data[data['total_sqft']/data['bhk'] >= 300]
```

```python
# Removing rows where bathrooms are more than bedroom xx
data = data[data['bath'] <= data['bhk']+2]
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories = [['Carpet  Area','Built-up  Area','Super built-up  Area', 'Plot  Area']] )
area = oe.fit_transform(data[['area_type']])
data['area'] = area.astype('int')
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['location'] = le.fit_transform(data['location'])
```

```python
data.drop('area_type',axis = 1,inplace = True)
```

```python
data.info()
```

### Feature Engineering

```python
x = data[['area','price_per_sqft', 'location', 'total_sqft', 'bath', 'bhk']]
y= data[['price']]
```

### Split Data into train and test

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 42)
```

### modelling

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
lr_r2 = r2_score(y_test,y_pred)
print(lr_r2)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
rf.fit(x_train, y_train)

```

```python
Ry_pred = rf.predict(x_test)
rf_r2 = r2_score(y_test,y_pred)
print(rf_r2)
```

```python
plt.scatter(y_test,Ry_pred,alpha = 0.5)
plt.show()
```
